```python
import pandas as pd
import numpy as np
df = pd.read_excel('Lab Session Data.xlsx',sheet_name=0)

df.set_index('Customer', inplace=True)
df
df=df.iloc[0:10,0:4]
Candies= df["Candies (#)"]
Mangoes = df["Mangoes (Kg)"]
Milk_Packets = df["Milk Packets (#)"]
Payment = df["Payment (Rs)"]
A = np.column_stack((Candies, Mangoes, Milk_Packets))
C = Payment.values
pseudo_inv = np.linalg.pinv(A)
X = pseudo_inv.dot(C)
rank_A = np.linalg.matrix_rank(A)
print(A)
print("The dimensionality of the vector space is the rank of the matrix")
print("The number of vectors in this vector space are ",rank_A)
print("Rank of A:", rank_A)
print("The cost of each product are ",X,"respectively.")
```

```python
def encode_thyroid_data(df):
    binary_mapping = {'f': 0, 't': 1}
    binary_columns = [
        'sex', 'on thyroxine', 'query on thyroxine', 'on antithyroid medication',
        'sick', 'pregnant', 'thyroid surgery', 'I131 treatment','hypopituitary', 'psych',
        'TT4 measured','TSH','T3','TT4','T4U','FTI', 'T4U measured', 'FTI measured', 'TBG measured','TBG','referral source','Condition'
    ]

    for column in binary_columns:
        df[column] = df[column].map(binary_mapping)
        df[column] = df[column].astype('category').cat.codes




    return df
```

```python
thyroid_new=encode_thyroid_data(thyroid)
```

```python
thyroid_new

```

```python
thyroid_new.isnull()
```

```python
thyroid_new.replace('?',pd.NA,inplace=True)
```
