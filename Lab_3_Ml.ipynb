```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score
```

```python
df = pd.read_csv('DCT_withoutduplicate 8.csv')
```

```python
#mentioned only 2 classes needed
selected_classes = [3333,3334]
f_df = df[df['LABEL'].isin(selected_classes)]
f_df
X = f_df.iloc[:, :-1].values
y = f_df.iloc[:, -1].values
selected_classes_1 = [3333]

# Filter the DataFrame
f1_df = df[df['LABEL'].isin(selected_classes_1)]
print(f1_df)

selected_classes_2= [3334]

# Filter the DataFrame
f2_df = df[df['LABEL'].isin(selected_classes_2)]
f2_df

```

```python
#calculating mean for first class
f1_mean=[]
for col in f1_df.columns:
    mean_value = f1_df[col].mean(axis=0)
    f1_mean.append(mean_value)
f1_mean
#calculating mean for second class
f2_mean=[]
for col in f2_df.columns:
    mean_value = f2_df[col].mean(axis=0)
    f2_mean.append(mean_value)
f2_mean
```

```python
#calculating standard deviation for first class
f1_std=[]
for col in f1_df.columns:
    std_v = f1_df[col].std(axis=0)
    f1_std.append(std_v)
f1_std
#calculating standard deviation for second class
f2_std=[]
for col in f2_df.columns:
    std_v = f2_df[col].std(axis=0)
    f2_std.append(std_v)
f2_std
```

```python
#calculating distance between mean vecotrs between classes
distance = np.linalg.norm(np.array(f1_mean) - np.array(f2_mean))
print(distance)
```

```python
#splitting dataset to training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
#performing KNN classifier for different K values
k_values = range(1, 12)
accuracies = []


for k in k_values:

    knn = KNeighborsClassifier(n_neighbors=k)

    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
```

```python
# Plot the accuracy vs. k
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o')
plt.title('Accuracy vs. k in k-NN Classification')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
# Initialize lists to store precision, recall, and F1 score values
precisions = []
recalls = []
f1_scores = []

# Loop over different values of k
for k in k_values:
    # Initialize k-NN classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k)

    # Fit the model on the training data
    knn.fit(X_train, y_train)

    # Predict on the test data
    y_pred = knn.predict(X_test)

    # Calculate precision, recall, and F1 score
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Store the metrics
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)

    # Optionally, you can also print the classification report
    print(f"Classification Report for k={k}:\n")
    print(classification_report(y_test, y_pred, digits=4))
```

```python
plt.figure(figsize=(14, 8))

# Plot Precision
plt.plot(k_values, precisions, marker='o', label='Precision', color='blue')

# Plot Recall
plt.plot(k_values, recalls, marker='o', label='Recall', color='green')

# Plot F1 Score
plt.plot(k_values, f1_scores, marker='o', label='F1 Score', color='red')

plt.title('Precision, Recall, and F1 Score vs. k in k-NN Classification')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```
